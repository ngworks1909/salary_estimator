```python
import pandas as pd
```

```python
dataset = pd.read_csv('salary_dataset.csv')
```

```python
dataset.head()
```

```python
dataset = pd.get_dummies(dataset, columns=['Skill'])
x = dataset.drop('Salary', axis=1)
y = dataset.iloc[:, -1].values
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(x_train, y_train)
```

```python
score = regressor.score(x_test, y_test)
score * 100
```

```python
# ------------------------
# Custom Prediction Section
# ------------------------

input_skills = [("Python", 3), ("SQL", 2), ("Machine Learning", 1)]

total_salary = 0
print("🔍 Individual Skill Predictions:")

for skill, exp in input_skills:
    # Prepare input vector
    input_dict = {col: 0 for col in x.columns}
    input_dict['Experience'] = exp

    skill_col = f"Skill_{skill}"
    if skill_col in input_dict:
        input_dict[skill_col] = 1
    else:
        print(f"⚠️ Skill '{skill}' not found in training data.")
        continue

    # Predict
    input_df = pd.DataFrame([input_dict])
    predicted = regressor.predict(input_df)[0]
    total_salary += predicted

    print(f"💡 {skill} ({exp} years): ₹ {predicted:.2f} LPA")

# Final combined salary
print(f"\n💰 Total Combined Predicted Salary: ₹ {total_salary:.2f} LPA")

```
